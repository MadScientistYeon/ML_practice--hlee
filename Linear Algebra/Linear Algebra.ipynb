```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) #start, finish, n points 
#Return evenly spaced numbers over a specified interval

#Example
#np.linspace(2.0, 3.0, num=5)
#array([2.  , 2.25, 2.5 , 2.75, 3.  ])
#np.linspace(2.0, 3.0, num=5, endpoint=False)
#array([2. ,  2.2,  2.4,  2.6,  2.8])
#np.linspace(2.0, 3.0, num=5, retstep=True)
#(array([2.  ,  2.25,  2.5 ,  2.75,  3.  ]), 0.25)
```

Distance travelled by robber: d = 2.5t

```python
distance_robber = 2.5 * t
```

Distance travelled by sheriff: d = 3(t-5)

```python
distance_sheriff = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('A Bank Robber Caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40]) #x range in plot
ax.set_ylim([0, 100]) #y range in plot
ax.plot(t, distance_robber , c='green')
ax.plot(t, distance_sheriff, c='brown')
plt.axvline(x=30, color='blue', linestyle='--') #lines to indicate answer
_ = plt.axhline(y=75, color='blue', linestyle='--')
```
